# 1. Introdução a Finanças Quantitativas

Finanças quantitativas referem-se à aplicação de métodos **matemáticos** e **estatísticos**, com o **apoio da tecnologia**, para analisar dados financeiros e tomar decisões de investimento. Utiliza de modelos e algoritmos para entender e tentar prever movimentos do mercado, gerenciar riscos e/ou otimizar estratégias de investimento.

####**Utilização de Finanças Quantitativas:**

As finanças quantitativas são amplamente utilizadas no mundo financeiro para tomada de decisões e redução na incerteza associada aos mercados/investimentos.

Essa abordagem é aplicada em diversos contextos, incluindo **precificação de ativos, gestão de portfólio, modelagem de riscos** e derivativos, entre outros.

####**Vantagens das Finanças Quantitativas:**

*   **Precisão Analítica:** A utilização de métodos quantitativos permite uma análise mais precisa e objetiva dos dados financeiros, proporcionando uma compreensão mais profunda dos mercados.

*   **Gestão de Riscos:** As finanças quantitativas são essenciais para modelar e gerenciar riscos financeiros. Isso ajuda os investidores a tomar decisões mais informadas e a proteger seus investimentos contra movimentos adversos do mercado.

*   **Otimização de Portfólio:** Por meio de modelos matemáticos avançados, é possível otimizar a alocação de ativos em um portfólio, buscando maximizar o retorno esperado para um determinado nível de risco.

*   **Tomada de Decisão Baseada em Dados:** Ao utilizar abordagens quantitativas, as decisões financeiras são fundamentadas em dados históricos e **estatísticos**, proporcionando uma base sólida para a tomada de decisão.

*   **Eficiência Operacional:** O uso de algoritmos e automação nas finanças quantitativas pode aumentar a eficiência na execução de estratégias de investimento, reduzindo a dependência de análises subjetivas.

*   **Adaptação Dinâmica:** Modelos quantitativos podem ser ajustados dinamicamente com base em novos dados, permitindo uma adaptação mais rápida às mudanças nas condições do mercado.

Resumindo, finanças quantitativas oferecem uma abordagem sistemática e baseada em dados para compreender, analisar e gerenciar questões financeiras, contribuindo para a tomada de decisões mais informadas e eficazes no mundo dos investimentos.

## 4.2. Médias móveis

As médias móveis são comumente utilizadas no mercado financeiro para analisar padrões e tendências em séries temporais de preços.

Elas representam a média de um conjunto de valores **em um determinado intervalo de tempo (janela)** e são recalculadas continuamente à medida que novos dados se tornam disponíveis.


> **As médias móveis podem ser usadas para:**
>
> *   **Identificar Tendências**: Cruzamentos entre médias móveis de curto e longo prazo podem indicar mudanças nas tendências de preços.
>
> *   **Sinalizar Pontos de Entrada e Saída**: Cruzamentos entre o preço e a média móvel podem ser usados como sinais de compra ou venda.
>
> *   **Estabelecer Níveis de Suporte e Resistência**: As médias móveis também são usadas para determinar níveis de suporte e resistência, indicando áreas onde o preço pode encontrar dificuldade em se mover além.


Resumindo: médias móveis são uma ferramenta versátil que nos fornece insights  sobre a direção e a força das tendências nos mercados financeiros, sendo amplamente utilizadas por traders e analistas técnicos para tomar decisões.

No caso utilizamos a função de agregação .mean() para calcular a média móvel.

## 4.3. Retornos diários

Os retornos diários são uma medida comum utilizada em finanças para avaliar o desempenho de um ativo ao longo de um único dia de negociação.

Representa a variação percentual no valor de um ativo de um dia para o outro.

Calculo dos retornos diários com base nos valores de Adj Close de PETR

```python
retornos = petro['Adj Close'].pct_change()
retornos
```

```python
retornos.plot();
```

#### **Retorno diário médio**


O Retorno Diário Médio, em inglês "*Average Daily Return*", é a média aritmética dos retornos diários ao longo de um determinado período de tempo.

É uma medida útil para entender o desempenho médio de um ativo durante o período.

```python
# Calculando a média dos retornos diários com todos os dados que temos
retornos.mean()
```

#### **Média móvel dos retornos**


Juntando os dois conceitos estudados até agora, podemos então calcular a média móvel dos retornos de uma determinada janela.

```python
# Gráfico da Média Móvel de 50 períodos dos retornos diários
retornos.rolling(50).mean().plot();
```

## 4.7. Métricas de retorno acumulado

#### **Cumulative Return**

O Retorno Acumulado, em inglês "*Cumulative Return*" , é uma medida que **expressa o ganho total ou a perda** de um investimento ao longo de um determinado período de tempo.

```python
retorno_acumulado = (retornos + 1).cumprod()

retorno_acumulado.plot(figsize = (12,6));
```

Vamos calcular agora o cumulativo máximo, ou seja, os picos atingidos ao longo do tempo.

O conceito de "cumulativo máximo" refere-se ao valor mais alto alcançado por uma determinada métrica ao longo de um período de tempo.

Para isso usaremos o método `cummax`

```python
picos = retorno_acumulado.cummax()

picos.plot(figsize = (12,6));
```

```python
# Transformar os dados de retorno de pd.Series em um pd.DataFrame
retornos = pd.DataFrame(retornos)
```

```python
# Agora vamos criar novas colunas com os outros dados já calculados
retornos['Retorno Acumulado'] = retorno_acumulado
retornos['Picos'] = picos

retornos.head()
```

```python
# Com esse DataFrame em mãos podemos criar visualizações interessantes
# Exemplo Retorno Acumulado e seus picos
retornos[['Retorno Acumulado', 'Picos']].plot(figsize = (12,6));
```

#### **Drawdown**

Aprendido os conceitos anteriores, podemos avançar e entender o conceito de "*drawdown*", que se refere à **medida da queda de um investimento em relação ao seu ponto mais alto até o seu ponto mais baixo**.

Ou seja, é a redução percentual máxima do valor de um investimento desde o último pico até o próximo vale.

O drawdown é uma métrica crítica na avaliação de riscos, proporcionando insights sobre as perdas potenciais que um investidor pode enfrentar.

```python
drawdown = (retorno_acumulado - picos)/picos

drawdown.plot(figsize = (12,6));
```

Qual o drawdown máximo?

É justamente o valor mais 'fundo' que o gráfico acima atinge

```python
drawdown.min()
```

# 6. Estudo de caso: matriz de correlação

Para este estudo vamos utilizar vários dados: FIIS, ETFS, BDRs, Moedas, Cryptos e Índices.

## 6.1. Importando dados de FIIs, ETFs e BDRs

```python
lista_ativos = ['KNRI11.SA', 'ONEF11.SA', 'HCTR11.SA', 'HGLG11.SA',
          'HGBS11.SA', 'SMAL11.SA', 'IVVB11.SA', 'AAPL34.SA']

df_ativos = yf.download(lista_ativos, start = '2017-01-01')['Adj Close']
```

## 6.2. Importando dados de moedas, índices, e criptomoedas

```python
lista_cesta_moedas = ['USDBRL=X', 'EURBRL=X', '^BVSP','^GSPC', 'BTC-USD', 'ETH-USD']
df_cesta_moedas = yf.download(lista_cesta_moedas, start = '2017-01-01')['Adj Close']
```

```python
# Renomeando o nome das colunas
df_cesta_moedas.rename(columns = {'USDBRL=X': 'Dolar',
                                  'EURUSD=X': 'Euro',
                                  '^BVSP': 'Ibov',
                                  '^GSPC': 'S&P500',
                                  'BTC-USD': 'Bitcoin',
                                  'ETH-USD': 'Ethereum'}, inplace = True)
```

## 6.3. Importando dados de ações

```python
lista_acoes = ['PETR4.SA', 'WEGE3.SA', 'MGLU3.SA', 'LREN3.SA', 'VALE3.SA', 'BPAC11.SA','OIBR3.SA']

df_acoes = yf.download(lista_acoes, start = '2017-01-01')['Adj Close']
```

## 6.4. Juntando tudo num dataframe só

Para isso vamos utilizar uma função  nova chamada `reduce`.

Ela faz parte do módulo `functools` e é **utilizada para aplicar uma função cumulativa a todos os itens de uma sequência**, de maneira a reduzir a sequência a um único valor.

A sintaxe básica da função reduce é a seguinte:

`functools.reduce(função, sequência[, inicial])`

**função:** A função cumulativa que será aplicada repetidamente aos itens da sequência.

**sequência:** A sequência de itens a ser reduzida.

inicial: (Opcional) Um valor inicial que será utilizado como primeiro argumento na primeira chamada da função. Se não fornecido, o primeiro e o segundo itens da sequência serão utilizados.

```python
# Importando a biblioteca
from functools import reduce
```

```python
dfs = [df_ativos, df_cesta_moedas, df_acoes]
```

```python
# Aplicando a função reduce para fazer o merge dos 3 DataFrames
df_final = reduce(lambda left, right: pd.merge(left, right, how = 'inner', on = 'Date'),dfs)

# Resultado do merge de todos os DFs
df_final.head()
```

## 6.5. Preparar os dados para a criação da matriz de correlação

```python
# Calculando o retorno diário do df_final e excluindo os valores nulos
retornos = df_final.pct_change().dropna()

retornos
```

## 6.6. Matriz de correlação

```python
# Calculando a correlação do retorno de todos os ativos do df_final
correlacao = retornos.corr()

correlacao
```

Para uma melhor visualização da tabela, utilizamos o gráfico heatmap da biblioteca Seaborn.

```python
# importando a biblioteca seaborn
import seaborn as sns
```

```python
sns.heatmap(correlacao)
```

```python
# Podemos melhorar um pouco a visualização
# Marcando o parametro annot = True para ver os valores em cada quadrante
# E podemos mudar as cores passando um cmap

plt.figure(figsize = (16,8))

sns.heatmap(correlacao,
            vmin = -1, vmax = 1, annot = True, cmap = 'BrBG'
)
```

# 7. Estudo de caso: comparação de carteira com o IBOV

Por simplificação nós vamos simular um único aporte feito no primeiro dia

## 7.1. Importação dos dados

```python
# Lista de ativos da carteira
ativos = ['PETR4','VALE3', 'ABEV3', 'WEGE3', 'EMBR3', 'RADL3', 'CVCB3', 'CIEL3', 'OIBR3', 'KNRI11']

# Incluindo o .SA para conseguirmos puxar os dados do yfinance
ativos = [i + '.SA' for i in ativos]

# Baixando os dados da biblioteca yfinance
df = yf.download(ativos, start = '2021-01-01')['Adj Close']

# Visualizando
df.head()
```

## 7.2. Métricas básicas de carteira

Preços do dia do aporte

```python
# Os primeiros valores de cada ativo
preco_aporte = df.iloc[0]
```

Se investimos R$ 1000 em cada papel, quantos papéis precisamos comprar de cada?

```python
# Arredondando pois não podemos comprar valores fracionados
qtd_papeis = round(1000 / preco_aporte, 0)

qtd_papeis
```

## 7.3. Construção da carteira

Como saber qual era o valor que tínhamos em cada papel a cada dia?

Basta multiplicar a quantidade de papéis que compramos com os preços diários (df)

```python
carteira = df * qtd_papeis

carteira.head()
```

```python
# Na última linha temos o resultado de cada ativo
carteira.tail()
```

Criando uma nova coluna que vai demonstrar o valor total da nossa carteira, por dia

```python
carteira['PL Total'] = carteira.sum(axis = 1)

carteira
```

```python
carteira['PL Total'].plot();
```

## 7.4. Comparação com o IBOV

Vamos obter agora os dados do IBOV para compará-lo com nossa carteira

```python
# Baixando os dados de Close do Ibovespa do yfinance
ibov = yf.download('^BVSP', start = '2021-01-01')[['Close']]

# Renomeando o nome da coluna
ibov.rename(columns = {'Close':'IBOV'}, inplace = True)

ibov
```

Agora vamos juntar os dados das tabelas do IBOV e da nossa carteira

```python
df_comparativo = pd.merge(ibov, carteira, how = 'inner', on = 'Date')

df_comparativo
```

```python
# Não dá para compararmos visualmente os dados brutos
df_comparativo.plot();
```

Precisamos normalizar tudo para colocá-los na mesma base.

```python
# Normalizando na base 100 (basta multiplicar por 100 no final)
df_comparativo_norm = (df_comparativo / df_comparativo.iloc[0]) * 100
```

```python
# Agora conseguimos comparar
df_comparativo_norm.plot();
```

```python
# Filtrando apenas o PL TOTAL e o IBOV para o gráfico ficar menos poluído
df_comparativo_norm[['IBOV', 'PL Total']].plot(figsize = (14,7));
```
